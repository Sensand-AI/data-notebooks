This notebook is a demo showing ways of applying color to rasters and manipulating color ranges and legends.

```python
# !pip install matplotlib-scalebar

import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib_scalebar.scalebar import ScaleBar

import rasterio as rst
from rasterio.plot import show
```

```python
cmpi_dir = 'data/temp/images-for-color-and-legend/CMPI_clipped.tif'
cmpi=rst.open(cmpi_dir)
area_1 = cmpi.read()

raster_file = 'path_to_your_raster_file.tif'

with rst.open(cmpi_dir) as src:
    area_1 = src.read(1)  # Read the first band
    transform = src.transform

cmap = plt.cm.get_cmap('viridis', 10)

#the actual data value range for this tif is greater than -0.02/0.02. But we cap the colours.
norm = mcolors.Normalize(vmin=-0.02, vmax=0.02)

fig, ax = plt.subplots()  # Create a figure and an axes object
raster_plot = ax.imshow(area_1, cmap=cmap, norm=norm, extent=rst.plot.plotting_extent(src))

cbar = fig.colorbar(raster_plot, ax=ax, extend='both')
# Uncomment and adjust the line below if you need custom labels on your colorbar
#cbar.ax.set_yticklabels(['<0.03', '0.03', '...', '>0.03'])

plt.show()

```
