```python
from pystac_client import Client
```

```python
STACK_URL = 'https://earth-search.aws.element84.com/v1'

# retrieving the root catalog, or landing page, of the API
cat = Client.open(STACK_URL)
# display the entire root catalog
cat
```

```python
# get all collections from a STAC
# @see https://github.com/radiantearth/stac-spec/blob/master/catalog-spec/catalog-spec.md
collections = cat.get_collections()

# list each collection in the catalog
for collection in collections:
    print(collection)
```

```python
# get a specific collection i.e. sentinel-2-l2a

collection = cat.get_collection('sentinel-2-l2a')

collection
```

```python
# get all items for the chosen collection

items = collection.get_items()

# limit the number of items retrieved, get top 10
count = 0
for item in items:
    print(item)
    count = count + 1
    if count > 10:
        break
```

```python
# lets get the first item for the collection we chose
item = next(items)
item
```
