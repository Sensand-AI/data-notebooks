Notebook testing deployment of a roboflow ML model trained on a hosted image dataset.

The actual data is annotated and uploaded to roboflow, and then trained there. The model used in this case was the option for the 'free' tier of roboflow:
 - Model Type: Roboflow 3.0 Object Detection (Fast)
 - Checkpoint: COCOn


Once a model has bene trained it can be deployed using an endpoint Roboflow generates, which is what this notebook demo's.

It is important to note that Roboflow only does object detection - you still need to write the logic/script on top of that. For example, if you want to count the number of screws, you need to write that script in python. The roboflow model will detect the screws, and then your additional python script sums them to get the total.

Roboflow provides an API through which you can upload an image and retrieve predictions from your model. This API is available in the Roboflow Python SDK, REST API, and CLI.

```python
model = project.version("1").model

# optionally, change the confidence and overlap thresholds
# values are percentages
model.confidence = 50
model.overlap = 25

# predict on a local image
img_path = "notebooks/sandbox/data/input-data/roboflow/Screw-175-.jpg"

prediction = model.predict(img_path, hosted=True)
```
