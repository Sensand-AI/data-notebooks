```python
import os
import json
import pandas as pd
import numpy as np
import geopandas as gpd

from geodata_fetch import settingshandler, harvest
from IPython.display import display, JSON

import holoviews as hv
import geoviews as gv
import rioxarray
import panel as pn

hv.extension('bokeh')
gv.extension('bokeh')
pn.extension() 
```

```python
# Function to list all .tif files in the specified directory
def list_tif_files(path):
    return [f for f in os.listdir(path) if f.endswith('.tif')]

# Function to load and display the selected .tif file
def load_and_display_tif(filename):
    filepath = os.path.join(data_load_path, filename)
    img = gv.util.from_xarray(rioxarray.open_rasterio(filepath).rio.reproject('EPSG:3857'))
    
    # Define map tiles and create the map image
    map_tiles = gv.tile_sources.EsriImagery().opts(width=1000, height=600)
    map_img = gv.Image(img, kdims=['x', 'y']).opts(cmap='viridis', title=filename)
    map_combo = map_tiles * map_img
    
    return map_combo
```

```python
# notebooks/sandbox/data
input_directory = '/workspace/notebooks/sandbox/data/input-data/planfarm-04-2024'
output_directory = '/workspace/notebooks/sandbox/data/output-data/planfarm-04-2024'

#property name, which will be used for filenames
property_name = 'terrawise_01'

file_path = '/workspace/notebooks/sandbox/data/input-data/planfarm-04-2024/planfarm_04_2024_paddocks.geojson'
geom = gpd.read_file(file_path)
gpd_lon = geom.centroid.x[0]
gpd_lat = geom.centroid.y[0]


bbox = list(geom.total_bounds)
```

```python
geom.plot()
```

```python
test_buffer = geom.buffer(0.002, resolution =16, join_style=2)
test_buffer.plot()
```

```python
# Column names for latitude and longitude in input file:
colname_lat = gpd_lat
colname_lng = gpd_lon

add_buffer = True
resample = False

# Bounding box: Left (min Long), Bottom (min Lat), Right (max Long), Top (max Lat)
target_bbox = bbox
# Resolution of data download in arcseconds (1 arcsec ~ 30m)

target_proj = "EPSG:3857"
target_res = 3

date_start = "2023-01-01"
date_end = "2024-01-01"
# Number of time intervals to split the image collection into
time_intervals = 0

target_sources = {"DEM":"DEM",
                  "SLGA":{
                      "Bulk_Density": ["0-5cm"],
                      "Organic_Carbon": ["0-5cm","5-15cm", "15-30cm"],
                      "Clay": ["0-5cm"],
                      "Silt": ["0-5cm"],
                      "Sand": ["0-5cm"],
                      "PH_CaCl2": ["0-5cm"],
                      "Effective_Cation_Exchange_Capacity": ["0-5cm"],
                  },
                  "Radiometric":["radmap2019_grid_k_conc_filtered_awags_rad_2019",
                                 "radmap2019_grid_th_conc_filtered_awags_rad_2019",
                                 "radmap2019_grid_u_conc_filtered_awags_rad_2019"]}

data_mask=True
```

```python
# Conversion to desired JSON format
"""
TODO: add default date inside appropriate function in harvest.py instead of here.
"""
json_data = {
    "property_name":property_name,
    "outpath": output_directory,
    "data_mask": data_mask,
    "target_crs": target_proj,
    "target_res": str(target_res),
    "date_start": date_start if date_start is not None else "2022-10-01", #a date of some kind must be provided or the harvester complains
    "date_end": date_end if date_end is not None else "2022-11-30",
    "time_intervals": time_intervals,
    "target_sources": target_sources,
    "target_centroid_lat": colname_lat,
    "target_centroid_lng": colname_lng,
    "target_bbox": target_bbox,
    "add_buffer": add_buffer,
    "resample": resample
}
```

```python
# Store the JSON data in a file
json_outpath = output_directory
json_outpath_fname = property_name + '_' +'data_harvest_settings.json'

outfname = os.path.join(json_outpath,json_outpath_fname)

with open(outfname, "w", encoding='utf-8') as file:
    json.dump(json_data, file, ensure_ascii=False, indent=4)
```

```python
# Path to file:
path_settings = output_directory
# Filename
fname_settings = json_outpath_fname

infname = os.path.join(path_settings,fname_settings)
```

```python
# print settings as json
settingshandler.display_settings(infname)
```

```python
# store settings as namespace (easier to interact with)
"""
TODO: have hard-coded geom for now to get masking done but the settings.json should be modified to pass the originall polygon rather than the bbox
"""
settings = settingshandler.main(infname)
```

```python
df = harvest.run(infname, geom)
```

```python
# Create the dropdown menu
data_load_path = output_directory + '/masked-data'
tif_files = list_tif_files(data_load_path)

dropdown = pn.widgets.Select(name='Select a .tif file', options=tif_files)

# Panel DynamicMap to update the map based on the dropdown selection
@pn.depends(dropdown.param.value)
def update_map(selected_file):
    return load_and_display_tif(selected_file)

# Layout using Panel
layout = pn.Column(
    pn.Row(pn.Column('## Geodata file viewer', dropdown)),
    update_map
)

# Display the Panel dashboard
layout.servable()
```
