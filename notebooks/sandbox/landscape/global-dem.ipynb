This notebook is to explore alternative elevation models to replace the currently used Geoscience Australia DEM. One, because we want to offer elevation data globally, and two because the GA DEM is unreliable to access and we often get 502 errors.

Potential data source:

- Copernicus GLO-30 or GLO-90
- available via AWS open data program
- https://registry.opendata.aws/copernicus-dem/
- Also has  STAC endpoint that can be used to access the data, so it keeps with the same access patterns we use for Sentinel data and Digital Earth Australia data.
- https://copernicus-dem-90m-stac.s3.amazonaws.com/
- You can explore the datasets available for this STAC here: http://apps.sentinel-hub.com/eo-browser/
