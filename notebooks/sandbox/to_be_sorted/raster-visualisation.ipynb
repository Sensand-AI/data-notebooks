```python
import os
import json
from io import StringIO
import numpy as np

from geodata_fetch.harvest import DataHarvester

import geopandas as gpd
import rasterio
from rasterio.plot import show
from rasterio.enums import Resampling
import rioxarray

import hvplot.pandas
import hvplot.xarray
import geoviews as gv
import datashader as ds
import xarray as xr

import scipy as sp
import matplotlib
import matplotlib.pyplot as plt

import shapely

gv.extension('bokeh')
```

```python
output_dir = '/workspace/notebooks/sandbox/data/output-data/raster-viz'

gpd = gpd.read_file('/workspace/notebooks/sandbox/adhoc_client_work/nannup_truffles/map.geojson')
```

```python

bbox = gpd.total_bounds
geom = gpd
json_bbox = bbox.tolist()

gpd_lon = (bbox[0] + bbox[2]) / 2
gpd_lat = (bbox[1] + bbox[3]) / 2
```

```python

slga_layer = "Organic_Carbon"
slga_layer_depth = "0-5cm"

data_mask=True
add_buffer = False
resample=False

# Resolution of data download in arcseconds (1 arcsec ~ 30m)
target_res = 1
target_proj = "EPSG:3857" #find way to set this once at top of notebook and  have it feed through to everything else

date_start = "2022-10-01" #these aren't needed for SLGA but are expected in the package and throw error if excluded - need to fix
date_end = "2022-11-30"
time_intervals = 0 #same issue as above, need to fix in geodata package to be optional.

# layer depth should be in list format to avoid errors and allow for multiple depths later on.
target_sources = {
    "SLGA": {
        slga_layer: [slga_layer_depth]
    }
}

json_data = {
    "property_name":'rasterviz',
    "outpath": output_dir,
    "data_mask": data_mask,
    "target_res": str(target_res),
    "target_crs": target_proj,
    "date_start": date_start if date_start is not None else "2022-10-01", #a date of some kind must be provided or the harvester complains
    "date_end": date_end if date_end is not None else "2022-11-30",
    "target_centroid_lat": gpd_lat,
    "target_centroid_lng": gpd_lon,
    "time_intervals": time_intervals,
    "target_sources": target_sources,
    "target_bbox": json_bbox,
    "add_buffer": add_buffer,
    "resample": resample
}

# Converting dictionary to JSON formatted string
data = json.dumps(json_data)

# Create a file-like object from JSON string
json_file_like = StringIO(data)
```

```python
harvester = DataHarvester(json_file_like, geom)
harvester.run()
```

```python
filepath = '/workspace/notebooks/sandbox/data/output-data/raster-viz/SLGA_Organic_Carbon_0-5cm_rasterviz.tiff'
raster = rioxarray.open_rasterio(filepath).rio.reproject('EPSG:3857')
img = gv.util.from_xarray(raster)

map_tiles = gv.tile_sources.EsriImagery().opts(width=1000, height=600)
map_img = gv.Image(img, kdims=['x', 'y']).opts(cmap='viridis', title='original data')
map_combo = map_tiles * map_img

map_combo
```

```python
upscale_factor = 2

new_width = int(raster.rio.width * upscale_factor)
new_height = int(raster.rio.height * upscale_factor)

raster_upsampled = raster.rio.reproject(raster.rio.crs, shape=(new_height, new_width), resampling=Resampling.bilinear)
```

```python
img = gv.util.from_xarray(raster_upsampled)

map_tiles = gv.tile_sources.EsriImagery().opts(width=1000, height=600)
map_img = gv.Image(img, kdims=['x', 'y']).opts(cmap='viridis', title='Bilinear resampled data')
map_combo = map_tiles * map_img

map_combo
```

Steps so far:
- use dea tools to get contours from the raster. Set the number of contours using nump linear spacing between the `min` and `max`
- The contours are returned as `multipolylines` so we then need to process them into polygons
- for each `multipolyline` need to:
    - explode the `multipolyline` into individual `linestrings`
    - use either a zero distance buffer or a convex hull to get a polygon from the linestring
    - Because each `multipolyline` represented all the areas of the same value (e.g. organic carbon values) they probably then need to be grouped back into a `multipolygon` for each original `multipolyline`.
    - Also need to ensure the `z_value` is preserved in the final output
