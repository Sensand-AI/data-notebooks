```python
import rasterio
import rasterio.plot
import matplotlib.pyplot as plt

# Path to your TIFF file
tiff_path = 'output_tiff.tiff'

# Open the TIFF file
with rasterio.open(tiff_path) as src:
    # Read the first band (assuming your DEM or raster data is in the first band)
    data = src.read(1)
    
    # Set up the plot
    plt.figure(figsize=(10, 10))
    
    # Use rasterio's show method to plot the image
    rasterio.plot.show(data, transform=src.transform, cmap='terrain', ax=plt.gca())
    
    plt.title('TIFF Preview')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.show()
```
